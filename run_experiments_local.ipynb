Notebook to run experiments locally

```python
from src.dataset import KreuzerSkarkeDataset
dataset = KreuzerSkarkeDataset(load_projections=True, projections_file='original')
```

```python
from src.models.vanilla_nn import Vanilla_nn
from src.models.hartford import Hartford
from src.models.invariantmlp import InvariantMLP
from src.models.xgboost import XGboost
from src.models.vision_transformer import VisionT
MODELS = [
    Vanilla_nn,
    Hartford,
    InvariantMLP,
    XGboost,
    VisionT
    ]
```

```python
all_accuracies = []
for model in MODELS:
    model_instance = model(dataset)
    model_instance.train(num_epochs=1)
    this_acc = model_instance.get_accuracy()
    print(f'This accuracy: {this_acc}')
    all_accuracies += [this_acc]

print(f'All accuracies: {all_accuracies}')
```
